## Feature Selection

```python
 # Creating seperate dataframe for all independent variables
independent_variables = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal']
x = heart[independent_variables]

print(type(x))
print(x)

target_variables = 'target'
y = heart[target_variables]

print(type(y))
print(y)

```

```python
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense

def getNormalizationData(x):

    # Fit scaler on training data
    norm = MinMaxScaler().fit(x)

    # transform training data
    x_norm = norm.transform(x)
    x_train_norm = pd.DataFrame(x_norm, columns= x.columns.values)
    
    return x_train_norm;

```

```python
# normalize x
# Normalization typically means rescales the values into a range of [0,1]
x_train_norm = getNormalizationData(x);
x_train_norm.describe()
# x_train_norm.head()
```

```python
# Convert to numpy array
x_numpy = np.array(x_train_norm)
y_numpy = np.array(y)

print(type(x_numpy))
print(type(y_numpy))
```

## Train-Test Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_numpy, y_numpy, test_size = 0.15, random_state=42)
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.05, random_state=42)
```

```python
print(y_train)
print(x_train)
print(x_test)
print(y_test)
```

## Selecting Model

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV


model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C' : [1,10,20],
            'kernel': ['rbf', 'linear']
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear', multi_class='auto'),
        'params':{
            'C':[1,5,10]
        }
    },
    'random_forest':{
        'model':RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10]
        }
    }

}

scores = []
```

```python
for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=False)
    clf.fit(x_train, y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
df = pd.DataFrame(scores, columns=['model','best_score', 'best_params'])
print(df)
```
